```python
user_input = input().strip()

if user_input == "":
    print("No input provided")
else:
    num = int(user_input)

    if num % 2 == 1:
        print("Odd")
    elif 2 <= num <= 5:
        print("Even and between 2 and 5")
    elif 6 <= num <= 20:
        print("Even and between 6 and 20")
    else:
        print("Even and greater than 20")

```
