**TIMING PYTHON CODE**

```python
def f1(n):
    return [str(num) for num in range(n)]
f1(10)
```

```python

def f2(n):
    return list(map(str,range(n)))
f2(10)
```

```python

import time

st = time.time()
r = f1(20)
et = time.time()
elt = et - st
print(f"Elapsed time = {elt}")
print(f"Result = {r}")
```

```python

st = time.time()
r = f2(20)
et = time.time()
elt = et - st
print(f"Elapsed time = {elt}")
print(f"Result = {r}")
```

```python
import timeit

stmt = '''f1(20)'''
setup = '''def f1(n): return [str(num) for num in range(n)]'''

timeit.timeit(stmt,setup,number=1000)
```

```python

import timeit

stmt2 = '''f2(20)'''
setup2 = '''def f2(n): return list(map(str,range(n)))'''

timeit.timeit(stmt2,setup2,number=1000)
```

```python
%%timeit
f1(20)
```

```python

%%timeit
f2(20)
```

**UNZIPPING & ZIPPING FILES**

```python
f = open("pr.txt",'w+')
f.write('Practice file 1')
f.close()
```

```python

f = open("pr2.txt",'w+')
f.write('Practice file 2')
f.close()
```

```python
import zipfile

cf = zipfile.ZipFile('cf.zip','w')
cf.write("pr.txt",compress_type=zipfile.ZIP_DEFLATED)
cf.write('pr2.txt',compress_type=zipfile.ZIP_DEFLATED)
cf.close()
```

```python

zip_obj = zipfile.ZipFile('cf.zip','r')
zip_obj.extractall("extracted_content")
```
