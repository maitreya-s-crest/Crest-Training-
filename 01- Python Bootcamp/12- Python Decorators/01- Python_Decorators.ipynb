**RETURNING FUNCTION**

```python
def hello(name='Jose'):
    print("hello() function is executed")

    def greet():
        return "greet() inside hello()"
    def welcome():
        return "welcome inside hello"

    if name == 'Jose':
        return greet()
    else:
        return welcome()
```

```python
hello()
```

```python
h = hello('Markie')
```

```python
print(h)
```

```python
welcome()
```

```python
def cool():
    def sup_cool():
        return "I am Super Cool"

    return sup_cool
```

**PASSING FUNCTION AS ARGUMENT**

```python
def hello():
    return "Hi Markie"

def other (func):
    print('Other code would go here')
    print(func())
```

```python
other (hello)
```

```python
def new_decorator(func):

    def wrap_func():
        print("Code would be here, before executing the func")

        func()

        print("Code here will execute after the func()")
    
    return wrap_func

def func_need_decorator():
    print("This function is in need of a decorator")
```

```python
func_need_decorator()
```

```python
func_need_decorator = new_decorator(func_need_decorator)
```

```python
func_need_decorator()
```

```python
@new_decorator

def func_need_decorator():
    print("This function is in need of a decorator")

func_need_decorator()
```
