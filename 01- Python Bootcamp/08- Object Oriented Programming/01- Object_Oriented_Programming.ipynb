**OBJECT ATTRIBUTES AND METHODS**

```python
class dog():
    def __init__(self,breed,name,age):
        self.breed = breed
        self.name = name
        self.age = age

    def bark(self):
        print(f"Woof! My name is {self.name} and my age is {self.age}.")

d = dog(breed = 'Husky', name = 'Sammy', age = 10)
d.bark()
```

```python
class circle():
    pi = 3.14
    def __init__(self,r):
        self.r = r
    def area(self):
        return self.pi*self.r**2

radius = float(input("Enter the Radius of Circle:"))
c = circle(r = radius)
result = c.area()
print(f"The Area of Circle is {result}")
```

**INHERITANCE**

```python
class animal():
    def __init__(self):
        print("Animal Class Created.")
    def who(self):
        print("I am an Animal.")
    def eat(self):
        print("I am eating.")

a = animal()
a.who()
a.eat()
```

```python
class dog(animal):
    def __init__(self,name,age):
        self.name = name
        self.age = age
        print("dog Class created from Class animal.")
    def who(self):
        print(f"Hi! I am a Dog. My name is {self.name} and my age is {self.age}.")
    def bark(self):
        print(f"Woof!")

d = dog(name = 'Bruno', age = 10)
d.who()
d.bark()
```

**MAGIC DUNDER METHODS**

```python
class book():
    def __init__(self, name, author, pages):
        self.name = name
        self.author = author
        self.pages = pages
    def __str__(self):
        return f"{self.name} by {self.author}"
    def __len__(self):
        return self.pages
    def __del__(self):
        print("Object has been deleted.")

b = book("Python","Markie",200)
print(str(b))
print(len(b))
print(b)
```
