```python
def user_input():
    choice = 'Wrong'

    while choice.isdigit() == False:

        choice = input("Enter digit: ") 

    return choice

user_input()
```

**USER INTERACTION**

```python
def user_input():
    choice = 'Wrong'
    a_range = range(1,10)
    w_range = False
    
    while choice.isdigit() == False or w_range==False:

        choice = input("Enter a digit in range of 0 to 10: ") 

        if choice.isdigit() == False:
            print("It's not a digit")
        else:
            if int(choice) in a_range:
                w_range = True
            else:
                w_range = False

    return int(choice)

user_input()
```

```python
def user_input():
    choice = 'Wrong'
    a_range = range(1,10)
    w_range = False
    
    while choice not in ['1','2','3']:

        choice = input("Enter a digit in [1,2,3]: ") 

        if choice not in [1,2,3]:
            print("Invalid")
        
    return int(choice)

user_input()
```
