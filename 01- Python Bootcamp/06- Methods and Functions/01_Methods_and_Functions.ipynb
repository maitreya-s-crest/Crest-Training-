```python
def player_guess():
    guess = ''
    while guess not in ['0', '1', '2']:
        guess = input ("Pick a number: 0,1 or 2 ")

    return int(guess)
```

```python
player_guess()
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == '0': 
        print ("Correct!")
    else:
        print ("Wrong Guess")
        print(mylist)
```

```python
mylist = ['', 'O', '']

#SHUFFLE LIST
mixedup_list = shuffle_list(mylist)

#USER GUESS
guess = player_guess()

#CHECK GUESS
check_guess(mixedup_list, guess)

```

```python
# *args - Arguments 
# **kwargs - Keyword arguments 
```

```python
def myfunc(a,b):
    #return 5% of the sum of a and b 
    return sum ((a,b)) * 0.05
```

```python
myfunc(40,60)
```

```python
def myfunc(a=0,b=0,c=0,d=0,e=0):
    return sum((a,b,c,d,e))*.05

myfunc(40,60,20)
```

```python
def myfunc(*args):
    return sum(args)*.05

myfunc(40,60,20)
```

```python
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print(f"My favorite fruit is {kwargs['fruit']}")
    else:
        print("I dont like fruits")
    myfunc(fruit= 'pineapple')
```
