**GENERATORS IN PYTHON**

```python
def create_cubes(n):
    result = []
    for x in range(n):
        result.append(x**3)
    return result
```

```python
create_cubes(10)
```

```python
for x in create_cubes(10):
    print(x)
```

```python
def create_cubes(n):
    result = []
    for x in range(n):
        yield x ** 3
```

```python
create_cubes(10)
```

```python
print (create_cubes)
```

```python
def gen_fibon(n):

    a = 1 
    b = 1
    for i in range(n):
        yield a
        a,b = b, a + b 
```

```python
for number in gen_fibon(10):
    print(number)
```

```python
def gen_fibon(n):

    a = 1 
    b = 1
    output = []

    for i in range(n):
        output.append(a)
        yield a
        a,b = b, a + b 
    return output
```

```python
for number in gen_fibon(10):
    print(number)
```
