**Task: Program to Reverse a Linked List Using Stack**

Problem Statement:
Given a singly linked list, reverse the order of its elements using an auxiliary stack. This method leverages the LIFO (Last-In-First-Out) nature of stacks to reverse the list.

```python
class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next
        
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def push(self, new_data): 
        new_node = Node(new_data, self.head) 
        self.head = new_node

    def reverseList(self): 
        stk = []
        ptr = self.head 
        while ptr.next != None: 
            stk.append(ptr) 
            ptr = ptr.next
        self.head = ptr 
        while len(stk) != 0: 
            ptr.next = stk.pop() 
            ptr = ptr.next
        ptr.next = None

    def printList(self):
        curr = self.head
        while curr: 
            print(curr.data, end = " ") 
            curr = curr.next
```

```python

if __name__ == "__main__":
    linkedList = LinkedList() 
    linkedList.push(5) 
    linkedList.push(4) 
    linkedList.push(3) 
    linkedList.push(2) 
    linkedList.push(1) 
    linkedList.reverseList() 
    linkedList.printList()
```
