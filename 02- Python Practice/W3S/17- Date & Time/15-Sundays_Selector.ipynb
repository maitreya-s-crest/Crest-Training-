**15. Sundays Selector**

Problem Statement: Write a Python program to **find and list all Sundays in a given year**. This can be useful for scheduling events, creating calendars, or doing date-based data analysis.


```python
import datetime
```

```python
def find_sundays(year):
    sundays = []

    dt = datetime.date(year, 1, 1)

    while dt.year == year:
        if dt.weekday() == 6:
            sundays.append(dt)
        dt += datetime.timedelta(days=1)
    return sundays
```

```python
year = int(input("Enter the year: "))
```

```python
sundays = find_sundays(year)
```

```python
print(f"\nAll Sundays in {year}:")
for sunday in sundays:
    print(sunday.strftime("%d-%m-%Y"))
```
