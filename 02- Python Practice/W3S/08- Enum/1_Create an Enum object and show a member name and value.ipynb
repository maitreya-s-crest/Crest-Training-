**1. Create an Enum object and show a member name and value**

Problem Statement: Write a Python program to create an Enum object and display a member name and value.



```python
from enum import Enum
class Country(Enum):
    India = 91
    Finland = 358
    Norway = 47
    Belgium = 32
    Italy = 39
    Antarctica = 672
print('\nMember name: {}'.format(Country.Finland.name))
print('Member value: {}'.format(Country.Belgium.value))
```
