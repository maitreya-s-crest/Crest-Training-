**9. String Number Checker Lambda**

Problem Statement: Write a Python program to check whether a given string is a number or not using Lambda.



```python
is_num = lambda s: (
    True if isinstance(s, str) and s.strip() and _check_num(s) else False
)

def _check_num(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

print(is_num('26587'))
print(is_num('4.2365'))
print(is_num('-12547'))
print(is_num('00'))
print(is_num('A001'))
print(is_num('001'))
print(is_num('-16.4'))
print(is_num('-24587.11'))
```
