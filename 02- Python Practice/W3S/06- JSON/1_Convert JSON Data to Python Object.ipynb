**1. Convert JSON Data to Python Object**

Problem Statement: Write a Python program to convert JSON data to Python object.



```python
import json 
```

```python
json_obj = '{"Name" : "Maitreya" , "Company": "Crest Infosystems" , "Age" : 21}'
```

```python
python_obj = json.loads (json_obj)
print ("\nJSON data:")
print (python_obj)
print("\nName: ", python_obj["Name"])
print("Company: ", python_obj["Company"])
print ("Age:" , python_obj["Age"])
```
