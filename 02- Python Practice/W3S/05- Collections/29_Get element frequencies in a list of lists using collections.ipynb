**29. Get element frequencies in a list of lists using collections**

Problem Statement: Write a Python program to get the frequency of elements in a given list of lists. Use the collections module.

```python
from collections import Counter 
```

```python
from itertools import chain 
```

```python
nums = [
    [1, 4, 0, 9],
    [0, 6, 1, 2],
    [2, 9, 0, 4],
]
```

```python
print("Original list of lists:")

print(nums)

print("\nFrequency of the elements in the said list of lists:")

result = Counter(chain.from_iterable(nums))

print(result) 
```
