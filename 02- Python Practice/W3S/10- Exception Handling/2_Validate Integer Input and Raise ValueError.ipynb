**2. Validate Integer Input and Raise ValueError**

Problem Statement: Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer.

```python
def get_integer_input (prompt):
    try:
        value = int(input(prompt))
        return value 
    except ValueError:
        print ("Error: Not a valid Integer.")

n = get_integer_input("Input an Integer: ")

print ("Input value: ", n)
```
