**7. Count Uppercase and Lowercase Letters in a String**

Problem Statement: Write a Python function that accepts a string and counts the number of upper and lower case letters.

Sample String : 'The quick Brow Fox'

Expected Output :
No. of Upper case characters : 3
No. of Lower case Characters : 12

```python
s = "The quick Brow Fox"

upper_count = 0
lower_count = 0

for char in s:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1

print("Uppercase:", upper_count)
print("Lowercase:", lower_count)



```
