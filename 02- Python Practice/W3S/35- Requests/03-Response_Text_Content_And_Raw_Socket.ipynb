Problem Statement:
Write a Python program that sends a request to a web page and:
1. Prints the response text (HTML content).
2. Prints the response content (in bytes).
3. Gets and prints the raw socket response from the server.

```python
import requests
```

```python
res = requests.get('https://www.google.com/')
```

```python
print("Response text of https://google.com/")
print(res.text)
print("\n==============================================================================")
print("\nContent of the said url:")
print(res.content)
print("\n==============================================================================")
print("\nRaw data of the said url:")
r = requests.get('https://api.github.com/events', stream = True)
print(r.raw)
print(r.raw.read(15))
```
