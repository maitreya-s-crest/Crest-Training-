**Inner Class**

```python
class student():
    def __init__(self,name,roll_no):
        self.name = name
        self.roll_no = roll_no
        #print("i am getting called from inside")
        self.l1 = self.laptop()

    def show(self):
        print(self.name + "-" + str(self.roll_no))
        print(self.l1.show())

    class laptop():
        def __init__(self):
            print("I am an inner class!!!")

        def show(self):
            return f"Hp"


s1 = student("Jaydeep",90)
s1.show()
l2 = student.laptop()
```

**inheritance**

```python
class A():
    def operation_1(self):
        print("operation_1")
    def operation_2(self):
        print("operation_2")

class B(A):
    def operation_3(self):
        print("operation_3")
    def operation_4(self):
        print("operation_4")
```

**Multiple Inheritance**

```python
class A():
    def operation_1(self):
        print("operation_1")
    def operation_2(self):
        print("operation_2")

class B(A):
    def operation_3(self):
        print("operation_3")
    def operation_4(self):
        print("operation_4")
        
class C(B):
    def operation_5(self):
        print("operation_5")
    def operation_6(self):
        print("operation_6")
```

**Multilevel Inheritance**

```python
class A():
    def operation_1(self):
        print("operation_1")
    def operation_2(self):
        print("operation_2")

class B():
    def operation_3(self):
        print("operation_3")
    def operation_4(self):
        print("operation_4")
        
class C(A,B):
    def operation_5(self):
        print("operation_5")
    def operation_6(self):
        print("operation_6")
```

**POLYMORPHISM (DUCK TYPING)**

```python
class Dog:
    def speak(self):
        print("Woof!")

class Cat:
    def speak(self):
        print("Meow!")

class Human:
    def speak(self):
        print("Hello!")

def make_it_speak(entity):
    entity.speak()  # Doesn't care what type it is

# Duck typing in action:
make_it_speak(Dog())     # Output: Woof!
make_it_speak(Cat())     # Output: Meow!
make_it_speak(Human())   # Output: Hello!
```

**Operator Overloading**

```python
class im_1():
    def __init__(self,a):
        self.a = a

    def __add__(self,other):
        return self.a + other.a

    def __gt__(self,other):
        return self.a > other.a
        
    def __str__(self):
        return f"{self.a}"

im1 = im_1(2)
im2 = im_1(3)
print(im1+im2)
print(im2>im1)
print(im1)
```

**Method Overloading**

```python
def add(*a):
    sum = 0
    for i in a:
        sum += i
    return sum
print(add(1,2,3,4,5))
print(add(1,2,3,4))
```

```python
def add(a = None , b = None , c = None):
    if c != None:
        return a + b + c
    elif b != None:
        return a + b
    else:
        return a

print(add(1,2,3))
print(add(1,2))
print(add(1))
```
