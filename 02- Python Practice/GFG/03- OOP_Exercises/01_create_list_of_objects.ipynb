<font size="6">Task: Create a List of Objects in Python Class</font>


**Problem Statement:**
**In Python, creating a list of objects involves storing instances of a class in a list. This allows for efficient access, modification, and iteration over multiple objects. For example, with a Geeks class having attributes like name and roll, a list can be used to manage multiple student records.**

```python
class Geeks:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        
```

```python
list = [('Markie', 29), ('Aduuuu', 11)]
```

<font size="5">**Using for loop**</font>

Define a class with the required attributes.
Prepare the input data (e.g., list of tuples or dictionaries).
Initialize an empty list to store the objects.
Use a for loop to iterate over the input data.
In each iteration, create an object and append it to the list.

```python
a = []
for data in list:
    a.append(Geeks(data[0], data[1]))
for obj in a:
    print(obj.name, obj.roll, sep=' ')
```

<span style="font-size: 20px;">**List Comprehension**</span>

Define a class with the required attributes.
Prepare the input data (e.g., list of names and roll numbers).
Use list comprehension to create a list of class instances from the input data.
Store and access the objects from the list as needed.

```python
a = [Geeks(name, roll) for name, roll in list]
```

```python
for obj in a:
    print(obj.name, obj.roll, sep=' ')
```
