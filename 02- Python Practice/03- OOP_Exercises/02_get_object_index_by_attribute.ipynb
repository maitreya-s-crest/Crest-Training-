<span style="font-size: 22px;">Task: Get Index in the List of Objects by Attribute in Python</span>

Problem Statement:
Given a list of objects, find the index of the first object that matches a specific attribute value using Python.

Steps:
Use the enumerate() function to loop through the list while tracking index and object.
For each object, check if the desired attribute matches the target value.
If a match is found, return the corresponding index.
If no match is found after the loop, handle it appropriately (e.g., return -1 or raise an exception).

```python
class X:
    def __init__(self, val):
        self.val = val
        
```

```python
def getIndex(li, target):
    for index, x in enumerate(li):
        if x.val == target:
            return index
    return -1
```

```python
list = [1,2,3,4,5,6]
```

```python
a = [X(i) for i in list]
```

```python
print(getIndex(a,3))
```
